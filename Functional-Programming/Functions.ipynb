````python
def calculate_sum(x,y):
    sum = x + y
    return sum

calculate_sum(4,6)
````

````python
def calculate(n):

    if n < 1:
          return "Number must be positive"
    sum=0
    for i in range(1,n+1):
        sum += i ** 2 
    return sum
    
print(calculate(3))
````

````python
def add_corresponding_number(x,y):
    sum = []
    if(len(x) == len(y)):
        for i in range (len(x)):
            sum.append(x[i] + y[i])
        return sum
    else:
        print("ERROR : 1Lengths are not equal")    
    
    

add_corresponding_number([1,2,3],[4,5,6])    
````
