```python
def absolute(x):
    if 0 > x:
        y = x * -1
        return y
    else:
        return x

absolute(-4)  
```

```python
my_lists = [1,2,3,4,5]

def find_square(a):
    return a ** 2

result = map(find_square,my_lists)
print(result)
print(list(result))
print(type(result))
```

```python
def faranhite_to_celcious(f):
    def convert(f):
        celcious = (f-32)*5/9
        celcious_round = round(celcious,2)
        return celcious_round
    temperatures = map(convert,f)
    print(list(temperatures))

faranhite_to_celcious([32,32,31,29])
```

```python
def correspondind_list(x,y):
    def calculate(x,y):
        return x + y
    
    sum = list(map(calculate,x,y))
    return sum
    

print(correspondind_list([1,2,3],[4,5,6]))   

# can pass any number of iterable to map()
```

```python
def filter_odds(x):
     return x % 2 == 0
result = filter(filter_odds,[9,5,8,12,13])

print(list(result))

```

```python
list2 = [12,15,18,20,25,30,35]

def divisible_by_three(x):
    return (x % 3 == 0) and (x % 5 != 0)
result = filter(divisible_by_three,list2)
print(list(result))    
```
