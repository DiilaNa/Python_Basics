### Positional Arguments

```python
def sum(x,y):
    return x+y

sum(2,5)
```

### Default Arguments

```python
def sum(x,y=10):
    return x+y

sum(3,4)
# sum(3) # return 13 , get default as 10
# sum(3,4,5) # Generate an error
```

```python
def cal_total(price,discount=0,qty=1):
    total = price * qty
    discount_price = (total * discount/100)
    s = total - discount_price

    return s

print(cal_total(100,10))
print(cal_total(100))
print(cal_total(100,10,2))
```

### Keyword Arguments

```python
def add(a,b):
    return a + b

print(add(a=10,b=10))
print(add(a=5,b=7))

```

### Aribitary Arguments

```python
# Arbitary positional arguments

def arbitary_positional_args(*args):
    print(args)
    print(type(args))

print(arbitary_positional_args(2))
print(arbitary_positional_args(2,5))
print(arbitary_positional_args(2,4,6,4,4))
```

```python
def sum_numbers(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(sum_numbers(4,4))
print(sum_numbers(4))
print(sum_numbers(4,4,6,7))
```

```python
def average_positive(*numbers):
    total = 0
    count = 0

    for num in numbers:
        if num > 0 :
            total += num
            count += 1

        if(num<=0):
            return "Number must be postive"    
        
    return total/count

print(average_positive(4,5))
```

```python
# Arbitary keyword arguments

def arbitary_keyword_args(**kwargs):
    for keys in kwargs.keys():
          print(keys)
    for values in kwargs.values():
            print(values)

    for key,value in kwargs.items():
        print(f"{key} : {value}")
 
print(arbitary_keyword_args(name="Alice",age=22,city="NewYork"))
```
