```python
add = lambda x,y:x+y
print(add(2,3))

print((lambda x,y:x+y)(4,5))
```

```python
max = lambda x,y : x if x > y else y
print(max(4,6))   
```

```python
h = [2,3,4,5]

sq = lambda h : h**2
result = map(sq,h)

print(list(result))
```

```python
l = [3,4,5,6,9,15]

d = lambda l:l % 3 == 0 and l % 5 != 0

result1 = filter(d,l)
print(list(result1))
```
