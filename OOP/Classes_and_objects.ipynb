## Classes in Python 
- written in pascal case
- do not use underscore (camel case)

```python
class Parrot:
    name = "Namal"
    parrot = 0

parrot_1 = Parrot()
print(parrot_1.name)
```

- first parameter should be self in an instance method
- not a keyword / work like 'this' keyword in java

```python
class Player:
    score = 10 # class attribute
    def cal_score(self,score): # instance method
        s = self.score + score # local attribute
        return s

p = Player()
print(p.cal_score(2)) 
print(p.cal_score(50)) 
```

```python
class Car:
    wheels = 4

    def drive_to(self,y):
        print(f"Drive to {y} on {self.wheels} wheels ")

    def show_wheel_count(self):
        print(f"This car runs on {self.wheels} wheels")

c= Car()
c.drive_to("A")
c.show_wheel_count()
```

```python
class Car:
    wheels = 4

    def __init__(self,brand,color):
        self.brand = brand
        self.color = color

    def drive_to(self,destination):
        print(f"The {self.color} {self.brand} is driving to {destination}")

    def show_wheel_count(self):
        print(f"The {self.color} {self.brand} runs on {self.wheels} wheels")
    
c= Car("Toyota","Red")
c.drive_to("Colombo")
c.show_wheel_count()
```

```python
import math
class Circle:
    def __init__(self,r):
        self.r = r

    def cal_area(self):
        self.area = math.pi * (self.r ** 2)
        return self.area
    
    def cal_circumference(self):
        self.circumference = 2 * math.pi * self.r
        return self.circumference
    
    def display(self):
        print(f"The area is {self.cal_area()}")
        print(f"The circumference is {self.cal_circumference()}")

c = Circle(7)
c.display()
```

```python
class Animal:
    def __init__(self,name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} speaks")
    
class Dog(Animal):
    def __init__(self,name,color):
        super().__init__(name)
        self.color = color

    def speak(self):
        print(f"{self.color} {self.name} speaks")

d = Dog("Tommy","Red")
d.speak()
```
